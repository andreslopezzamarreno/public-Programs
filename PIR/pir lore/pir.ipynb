```python
import pandas as pd

# Lee el contenido del archivo de texto
with open('archivo_limpio.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()

# Divide cada línea en sus campos usando el separador "|" y maneja campos vacíos
data = [line.strip().split('|') for line in lines]

# Asegúrate de que cada fila tenga la misma cantidad de campos
max_fields = max(len(row) for row in data)
for row in data:
    if len(row) < max_fields:
        row.extend([''] * (max_fields - len(row)))

# Crea un DataFrame con los datos y especifica los nombres de las columnas
df = pd.DataFrame(data, columns=["N.EXP","APELLIDOS Y NOMBRE","DNI","PAIS","SIT","VAL.","ERR.","NC.","VAL","PUNTUACION","BAREMO","PUNTUACION","PUNT. TOTAL","Nº ORDEN","T.","extra"])

#eliminar extra
df.drop('extra', axis=1, inplace=True)
df
```

```python
import pandas as pd

# Lee las primeras 150 líneas del archivo de texto
with open('archivo_limpio.txt', 'r', encoding='utf-8') as f:
    lines = [next(f).strip() for _ in range(150)]

# Divide cada línea en sus campos usando el separador "|" y maneja campos vacíos
data = [line.split('|') for line in lines]

# Asegúrate de que cada fila tenga la misma cantidad de campos
max_fields = max(len(row) for row in data)
for row in data:
    if len(row) < max_fields:
        row.extend([''] * (max_fields - len(row)))

# Crea un DataFrame con los datos y especifica los nombres de las columnas
column_names = ["N.EXP", "APELLIDOS Y NOMBRE", "DNI", "PAIS", "SIT", "VAL.", "ERR.", "NC.", "VAL", "PUNTUACION", "BAREMO", "PUNTUACION", "PUNT. TOTAL", "Nº ORDEN", "T.", "extra"]
df = pd.DataFrame(data, columns=column_names)

# Eliminar la columna extra si existe
if 'extra' in df.columns:
    df.drop('extra', axis=1, inplace=True)

df
```

```python
# Lista de columnas que deseas eliminar
columnas_a_eliminar = ['APELLIDOS Y NOMBRE', 'DNI','PAIS','SIT']

# Eliminar las columnas especificadas
df_limpio = df.drop(columns=columnas_a_eliminar)
df_limpio
```

```python
import pandas as pd
from natsort import natsorted

# Ordenar el DataFrame por la columna 'A' alfanuméricamente
df_ordenado = df_limpio.iloc[natsorted(df.index, key=lambda x: df.loc[x, 'Nº ORDEN'])]
df_ordenado
```
